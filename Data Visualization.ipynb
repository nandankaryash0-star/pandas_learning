```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
netflix = pd.read_csv('netflix_titles.csv')
netflix.head(5)
```

```python
#Number of movies vs TV shows (bar chart)
type_of_show = netflix['type'].value_counts()
type_of_show.plot(kind='bar')

plt.title("Movies vs TV show in Netflix")
plt.xlabel('Content type')
plt.ylabel('Total Count')

plt.xticks(rotation = 0)
plt.show()
```

```python
type_of_show = netflix.groupby('type').size()
type_of_show.plot(kind='pie',autopct = '%1.1f%%')

```

```python
#Content added over years (line chart)

over_year_growth = netflix.groupby('release_year').size()
over_year_growth.plot(kind='line')

plt.xlabel('Released Year')
plt.ylabel('Number of title')

plt.show()
```

```python
#Top 10 countries producing content (bar chart) 
performing_countries = netflix.groupby('country').size().sort_values(ascending=False).head(10)

performing_countries.plot(kind='bar')

plt.title("Top Performing Countries")
plt.xlabel('Countries')
plt.ylabel('No. of Shows Released')
```

```python
# This is a correct use of a histogram
netflix['release_year'].plot(kind='hist', bins=30, color='orange', edgecolor='white')
plt.title('Distribution of Release Years')
plt.show()
```
