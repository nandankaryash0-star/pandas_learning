```python
import pandas as pd
import numpy as np
```

```python
retail_store = pd.read_excel('OnlineRetail.xlsx')
```

```python
retail_store
```

```python
#Split data → Apply logic → Combine results (groupby)
retail_store.head()
```

```python
#Creating a revenue column

retail_store['Revenue'] = retail_store['Quantity'] * retail_store['UnitPrice']
retail_store
```

```python
retail_store['Revenue'].sum()
```

```python
#Best-Selling Products
retail_store.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
```

```python
#Top Revenue Products

retail_store.groupby('Description')['Revenue'].sum().sort_values(ascending=False).head(10)
```

```python
#Revenue by Country
retail_store.groupby('Country')['Revenue'].sum().sort_values(ascending=False).head(10)
```

```python
retail_store.isnull().sum()
```

```python
retail_store.duplicated().sum()
```

```python
retail_store.info()
```

```python
#Find top 5 customers by spending
retail_store.groupby('CustomerID')['Revenue'].sum().sort_values(ascending=False).head(5)
```

```python
#Find most frequent buyers

retail_store['CustomerID'].value_counts().head(5)
```

```python
#Find average order value

retail_store.groupby('InvoiceNo')['Revenue'].mean()
```

```python
retail_store['InvoiceDate'] = pd.to_datetime(retail_store['InvoiceDate'])

retail_store['Month'] = retail_store['InvoiceDate'].dt.month_name()

retail_store.groupby('Month')['Revenue'].sum().sort_values(ascending=False)
```

```python
#What is total revenue?

retail_store['Revenue'].sum()
```

```python
#Which product sells the most?
retail_store.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
```

```python
#Which country generates most revenue?

retail_store.groupby('Country')['Revenue'].sum().sort_values(ascending=False)
```

```python
#What are monthly sales trends?

retail_store.groupby('Month')['Revenue'].sum().sort_values(ascending=False)
```

```python
#Find average order value

retail_store.groupby('InvoiceNo')['Revenue'].sum().mean()
```

```python
retail_store['Date'] = retail_store['InvoiceDate'].dt.date
retail_store.groupby('Date')["Revenue"].sum().sort_values(ascending=False).head(1)
```

```python
import matplotlib.pyplot as plt
```

```python
#Monthly Revenue Trend
monthly_sales = retail_store.groupby("Month")["Revenue"].sum()

monthly_sales.plot()
plt.show()
```

```python
#Make It Look Good (VERY IMPORTANT)

monthly_sales.plot()

plt.title("Monthly Reveneu Trend")
plt.xlabel("Month")
plt.ylabel('Revenue')

plt.xticks(rotation=45)

plt.show()
```

```python
#Pie Chart (Proportions)
reveneue_by_country = retail_store.groupby('Country')['Revenue'].sum().head(5)

reveneue_by_country.plot(kind='pie',autopct='%1.1f%%')

plt.show()
```

```python
#Histogram (Distribution)
retail_store['Revenue'].plot(kind='hist',bins = 40)

plt.show()
```
